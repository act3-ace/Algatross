```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

n_points = 20

r = np.linspace(0, 1, n_points)
theta = np.linspace(0, 2*np.pi, n_points)
rp, thetap = np.meshgrid(r, theta)

R = Rotation.from_euler("zyx", [90, -45, 25], degrees=True)

x = np.reshape(rp * np.cos(thetap), (n_points**2,))
y = np.reshape(rp * np.sin(thetap), (n_points**2,))
z = np.reshape(rp, (n_points**2,))
polygon_rot = R.apply(np.stack([x, y, z]).T).T.reshape((3, n_points, n_points))

ax.plot_surface(polygon_rot[0], polygon_rot[1], polygon_rot[2], alpha=0.75)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
# ax.set_title('Basic 3D Cone')

plt.show()
```
